```python
import argparse
import os
from pathlib import Path
from typing import Optional
from penzai import pz
import jax.numpy as jnp
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
model1 = AutoModelForCausalLM.from_pretrained("laura000/Gemma-Instruct-7b-pruned_math_qa", torch_dtype=torch.float16, device_map = "auto")
tokenizer1 = AutoTokenizer.from_pretrained("laura000/Gemma-Instruct-7b-pruned_math_qa")
print(model1)
```

```python
params_model1 = model1.state_dict()

# Define the prefix of the parameters of layer 12
layer_12_prefix = 'model.layers.12.self_attn.q_proj.weight'  # This might vary based on your model architecture

# Filter the parameters of layer 12
layer_12_params_q = params_model1[layer_12_prefix]

print(layer_12_params_q)

print("PRINTING REPR:")

print(repr(layer_12_params_q))

array_to_render = layer_12_params_q

# Assuming q_proj_weight is the PyTorch tensor you want to convert
q_proj_weight_jax = jnp.array(layer_12_params_q.cpu().numpy())

# Now q_proj_weight_jax contains the JAX array corresponding to q_proj_weight
print(q_proj_weight_jax)

# Render the array
pz.ts.render_array(q_proj_weight_jax, truncate=True)
```
